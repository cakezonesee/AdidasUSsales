```python
import pandas as pd
import datetime as dt
```

```python
# Read the original Dataset without the header
sales = pd.read_excel('Adidas US Sales.xlsx', skiprows = 4).copy()
sales = sales.drop(columns = 'Unnamed: 0')
sales
```

# Create a new DataFrame

### which shows the avergae of "Price per Unit" and "Operating Maring" and the sum of "Unit Sold" and "Total Sales" in each "Year" in each "Region" and by each "Sales Method"

```python
# As we want to prioritize using only "Year" from dateTime "Invoice Date" so we will create a new column called "Year"
# and the value in this column will comes from the year value from dateTime "Invoice Date"

sales.insert(loc = 2, column = 'Year', value = sales['Invoice Date'].dt.year)
sales
```

```python
# Group the data together by "Year", "Region" and "Sales Method"

group = sales.groupby(['Year', 'Region', 'Sales Method'])
```

```python
# Calculate "average" of  "Price per Unit" and "Operating Margin" and sum of "Units Sold" and "Total Sales"
# Then put it inside "calculate" variable as new DataFrame

calculate = group.agg({ 'Price per Unit': 'mean',
                   'Units Sold': 'sum',
                   'Total Sales': 'sum',
                   'Operating Margin': 'mean'    
}).round(2)
calculate
```

```python
# Reset index for easier opration in the future

calculate = calculate.reset_index()
```

```python
calculate
```

```python
# Rename columns "Price per Unit" and "Operating Margin" to "Average Price per Unit" and "Average Operating Margin"
calculate = calculate.rename(mapper = {'Price per Unit': 'Average Price per Unit',
                                              'Operating Margin': 'Average Operating Margin'
                          }, axis = 1)
calculate
```

```python
with pd.ExcelWriter('Adidas Us Sales altered.xlsx') as writer:
    calculate.to_excel(writer, sheet_name = 'By Year, Region and Method', index = False)
```
